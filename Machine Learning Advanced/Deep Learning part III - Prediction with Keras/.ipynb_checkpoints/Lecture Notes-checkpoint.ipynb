# Um pouco de teoria sobre sklearn

## StandardScaler

O `StandardScaler`  uma tcnica de pr-processamento de dados que visa padronizar as features (variveis) de um conjunto de dados. A ideia  transformar as features para que elas tenham mdia zero e desvio padro igual a um. Isso  importante porque muitos algoritmos de Machine Learning assumem que as features esto em uma escala similar e tm distribuio normal ou aproximadamente normal.

Aqui est como voc pode usar o `StandardScaler` no sklearn:

```python
from sklearn.preprocessing import StandardScaler

# Crie uma instncia do StandardScaler
scaler = StandardScaler()

# Ajuste o scaler aos dados de treinamento
scaler.fit(X_train)

# Aplique a transformao aos dados de treinamento
X_train_scaled = scaler.transform(X_train)

# Aplique a mesma transformao aos dados de teste
X_test_scaled = scaler.transform(X_test)
```

Aqui est o que cada passo faz:

1. Importe a classe `StandardScaler` do mdulo `sklearn.preprocessing`.
2. Crie uma instncia do `StandardScaler` chamada `scaler`.
3. Ajuste o scaler aos dados de treinamento chamando o mtodo `fit(X_train)`, onde `X_train`  uma matriz (ou dataframe) contendo as features de treinamento.
   - Durante esse processo, o scaler calcula a mdia e o desvio padro de cada feature.
4. Aplique a transformao aos dados de treinamento chamando o mtodo `transform(X_train)`.
   - Isso ajusta as features de treinamento para que tenham mdia zero e desvio padro igual a um.
   - O resultado  armazenado na matriz `X_train_scaled`.
5. Aplique a mesma transformao aos dados de teste chamando o mtodo `transform(X_test)`.
   - Isso ajusta as features de teste usando a mesma mdia e desvio padro calculados a partir dos dados de treinamento.
   - O resultado  armazenado na matriz `X_test_scaled`.

Vamos ver um exemplo para entender melhor. Suponha que temos os seguintes dados de treinamento:

```
X_train = [[10, 2],
           [8, 3],
           [5, 1]]
```

E os seguintes dados de teste:

```
X_test = [[12, 4],
          [4, 0],
          [6, 2]]
```

Vamos aplicar o `StandardScaler`:

```python
from sklearn.preprocessing import StandardScaler

X_train = [[10, 2],
           [8, 3],
           [5, 1]]

X_test = [[12, 4],
          [4, 0],
          [6, 2]]

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Dados de treinamento originais:\n", X_train)
print("Dados de treinamento padronizados:\n", X_train_scaled)
print("Dados de teste originais:\n", X_test)
print("Dados de teste padronizados:\n", X_test_scaled)
```

A sada ser:
```python
# Dados de treinamento originais:
 [[10, 2], [8, 3], [5, 1]]
    
# Dados de treinamento padronizados:
 [[ 1.22474487 -0.26726124],
 [ 0.         1.33630621],
 [-1.22474487 -1.06904497]]
    
# Dados de teste originais:
 [[12, 4], [4, 0], [6, 2]]
    
# Dados de teste padronizados:
 [[ 2.44948974  1.87082869],
 [-0.81649658 -2.13808994],
 [-0.40824829 -0.26726124]]
```

Observe que os dados de treinamento originais foram transformados para ter mdia zero e desvio padro igual a um. As features foram ajustadas individualmente, ou seja, cada feature foi transformada separadamente com base na sua mdia e desvio padro. O mesmo ajuste foi aplicado aos dados de teste usando as mesmas mdias e desvios padro dos dados de treinamento.

A padronizao dos dados  til em vrios aspectos. Por exemplo, muitos algoritmos de Machine Learning, como regresso linear, regresso logstica e SVM, podem se beneficiar de features padronizadas, pois a escala das features no afetar o resultado do modelo. Alm disso, a padronizao pode ajudar a evitar que uma feature tenha mais influncia do que outras devido  escala.

Em resumo, o `StandardScaler`  uma tcnica de pr-processamento que padroniza as features, transformando-as para ter mdia zero e desvio padro igual a um. Isso  feito usando o mtodo `fit` para calcular as estatsticas de mdia e desvio padro a partir dos dados de treinamento e o mtodo `transform` para aplicar a transformao tanto nos dados de treinamento quanto nos dados de teste.

Sendo assim, vamos aplicar essas ideias dentro do nosso projeto:

# Regresso

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

regressor = Sequential()
````

````python
densidade = Dense(units = 1, 
                  input_dim = 1, 
                  kernel_initializer='Ones', 
                  activation = 'linear',
                  use_bias = True)
````

A funo `Dense()`  uma camada de neurnios em uma rede neural do Keras, uma biblioteca popular de Deep Learning. A funo Dense() cria uma camada totalmente conectada, em que cada neurnio da camada est conectado a todos os neurnios da camada anterior. Aqui esto alguns dos parmetros mais comuns que podemos passar para a funo `Dense()`:

- **`units`**: Especifica o nmero de neurnios (unidades) na camada.  um parmetro obrigatrio e deve ser um nmero inteiro positivo. Quanto maior o nmero de unidades, maior a capacidade da camada em aprender representaes complexas dos dados.

- **`input_dim`**: O parmetro input_dim  usado para especificar a dimenso (nmero de features) dos dados de entrada na primeira camada da rede neural quando se est construindo uma rede neural sequencial no Keras. O input_dim  usado em vez de input_shape quando se trabalha com uma rede neural sequencial simples, onde no h necessidade de especificar explicitamente a forma completa dos dados de entrada. Em vez disso, basta fornecer o nmero de features (dimenso) dos dados de entrada.

- **`kernel_initializer (A)`**: Especifica como os pesos da camada sero inicializados.  responsvel por definir os valores iniciais dos pesos dos neurnios. Alguns exemplos comuns de inicializadores de pesos so `'random_uniform'` (inicializao aleatria com distribuio uniforme), `'glorot_uniform'` (inicializao de Glorot com distribuio uniforme) e `'he_uniform'` (inicializao de He com distribuio uniforme). O valor padro  `'glorot_uniform'`.

- **`use_bias (B)`**: Um valor booleano que especifica se a camada usar bias ou no. O bias  til para adicionar flexibilidade aos neurnios e geralmente  ativado (True) por padro.

- **`activation (C)`**: Especifica a funo de ativao a ser usada nos neurnios da camada. A funo de ativao introduz no-linearidade na rede neural e ajuda a aprender relaes complexas nos dados. Alguns exemplos comuns de funes de ativao so `'relu'` (retificao linear), `'sigmoid'` (sigmoide), `'tanh'` (tangente hiperblica) e `'softmax'` (usada em problemas de classificao multiclasse). Se nenhum valor for especificado, nenhuma funo de ativao ser aplicada (camada linear).

- **`bias_initializer`**: Especifica como os bias da camada sero inicializados. O bias  um valor adicional adicionado aos neurnios que permite que a camada aprenda a funo desejada de maneira mais flexvel. Alguns exemplos comuns de inicializadores de bias so `'zeros'` (inicializao com valor zero), `'ones'` (inicializao com valor um) e `'random_normal'` (inicializao aleatria com distribuio normal). O valor padro  `'zeros'`.

- **`input_shape`**: Especifica a forma dos dados de entrada na primeira camada da rede neural.  necessrio fornecer `input_shape` apenas para a primeira camada da rede, pois as camadas subsequentes inferiro automaticamente a forma de entrada com base nas sadas das camadas anteriores.

Esses so apenas alguns dos parmetros mais comuns que podem ser passados para a funo `Dense()`. H tambm outros parmetros disponveis, como `kernel_regularizer` e `bias_regularizer`, que permitem regularizao dos pesos e bias da camada.

Vale ressaltar que a funo `Dense()`  usada em conjunto com outras camadas para construir uma rede neural completa, como uma rede neural sequencial ou uma rede neural com mltiplas entradas ou sadas.


$$\huge\displaystyle y = f(x) = A.x + B$$

$$\huge\displaystyle a(y) = C.y$$



````python
regressor.add(densidade)
````

````python
regressor.compile(loss='mean_squared_error', optimizer = 'adam')
````

Dentro do mtodo `compile()` do Keras, voc pode passar vrios parmetros que definem a configurao do processo de treinamento da sua rede neural. Aqui esto alguns dos parmetros mais comuns e sua funo:

- `optimizer`: Especifica o otimizador a ser usado durante o treinamento da rede neural. O otimizador  responsvel por atualizar os pesos da rede com base no erro calculado durante o treinamento. Alguns exemplos populares de otimizadores so `'adam'`, `'sgd'` (descida de gradiente estocstica) e `'rmsprop'`. Cada otimizador tem suas prprias configuraes e hiperparmetros adicionais que podem ser ajustados.

- `loss`: Especifica a funo de perda (loss function) a ser usada durante o treinamento. A funo de perda mede o quo bem o modelo est se saindo durante o treinamento, comparando as previses do modelo com os rtulos (targets) reais. A escolha da funo de perda depende do tipo de problema que voc est resolvendo. Por exemplo, para um problema de classificao binria, voc pode usar `'binary_crossentropy'`, e para um problema de regresso, voc pode usar `'mean_squared_error'`.

- `metrics`: Uma lista das mtricas a serem avaliadas durante o treinamento e teste do modelo. Essas mtricas so usadas para monitorar o desempenho do modelo durante o treinamento e podem ser teis para fins de avaliao. Alguns exemplos de mtricas comuns so `'accuracy'` (acurcia) para problemas de classificao e `'mean_squared_error'` para problemas de regresso.

- `loss_weights`: Um dicionrio ou uma lista que especifica os pesos atribudos a diferentes funes de perda (loss functions) quando a rede neural possui vrias sadas. Isso  til quando a rede neural tem vrias tarefas de aprendizado, e voc deseja dar mais importncia a uma tarefa especfica em relao s outras.

- `metrics_weights`: Um dicionrio ou uma lista que especifica os pesos atribudos a diferentes mtricas quando a rede neural possui vrias sadas. Isso  til quando voc deseja dar mais importncia a uma mtrica especfica em relao s outras ao avaliar o desempenho do modelo.

- `sample_weight_mode`: Especifica como as amostras so ponderadas durante o treinamento da rede neural. Isso  til quando voc deseja dar mais peso a algumas amostras em relao a outras, com base em algum critrio especfico.

- `weighted_metrics`: Uma lista de mtricas que sero usadas para calcular a mdia ponderada das mtricas durante o treinamento e teste do modelo. Isso  til quando voc deseja calcular mtricas com pesos diferentes, por exemplo, considerando a distribuio de classes desequilibrada.

Esses so apenas alguns dos parmetros mais comuns que podem ser passados para o mtodo `compile()` do Keras. Existem outros parmetros adicionais disponveis, dependendo das necessidades especficas do seu modelo, como `learning_rate`, `momentum`, `decay`, entre outros.  importante consultar a documentao do Keras para obter mais informaes sobre os parmetros disponveis e como us-los adequadamente para a sua

# Modelo: 'adam'

O 'adam'  um algoritmo de otimizao popular e eficiente usado em redes neurais. Ele  uma combinao do mtodo do gradiente descendente estocstico (SGD) com algumas ideias do mtodo de otimizao adaptativa (AdaGrad) e do algoritmo RMSprop.

A sigla 'adam' significa Adaptive Moment Estimation, que se refere s duas principais ideias incorporadas no algoritmo: adaptao dos passos do gradiente (moment) e adaptao da taxa de aprendizado (estimation).

O 'adam' utiliza um conceito chamado momento, que  uma mdia mvel dos gradientes anteriores. Ele armazena duas mdias mveis exponenciais: o momento do gradiente (momentum) e o momento do gradiente ao quadrado (momentum squared). Essas mdias mveis so atualizadas a cada iterao do treinamento.

As frmulas matemticas do 'adam' so as seguintes:

1. Clculo dos momentos do gradiente:

- Momento do gradiente:
 $$m_t = \beta_1 \cdot m_{\text{prev}} + (1 - \beta_1) \cdot g$$
- Momento do gradiente ao quadrado:
 $$v_t = \beta_2 \cdot v_{\text{prev}} + (1 - \beta_2) \cdot g^2$$
- Traduzindo para cdigo (Python):
```python
m_t = beta_1 * m_prev + (1 - beta_1) * g
v_t = beta_2 * v_prev + (1 - beta_2) * g^2
```

Onde:

- `m_t` ($m_t$)  o momento atualizado do gradiente.
- `v_t` ($v_t$)  o momento atualizado do gradiente ao quadrado.
- `g` ($g$)  o valor do gradiente para um dado peso (ou seja, a derivada parcial da funo de perda em relao ao peso).
- `beta_1` ($\beta_1$) e `beta_2` ($\beta_2$) so hiperparmetros que controlam a taxa de decaimento dos momentos.
- `m_prev` ($m_{\text{prev}}$) e `v_prev` ($v_{\text{prev}}$) so os momentos anteriores do gradiente.

<hr>

2. Correo dos momentos do gradiente para lidar com o vis inicial:

- Momento do gradiente corrigido:
 $$\hat{m}_t = \frac{m_t}{1 - \beta_1^t}$$
- Momento do gradiente ao quadrado corrigido:
 $$\hat{v}_t = \frac{v_t}{1 - \beta_2^t}$$
- Traduzindo para cdigo (Python):
```python
m_t_hat = m_t / (1 - beta_1^t)
v_t_hat = v_t / (1 - beta_2^t)
```

Onde:
- `m_t_hat` ($\hat{m}_t$) representa o momento do gradiente corrigido.
- `v_t_hat` ($\hat{v}_t$) representa o momento do gradiente ao quadrado corrigido.
- `t` ($t$)  o nmero de iteraes.

<hr>

3. Atualizao dos pesos:
- Pesos atualizados:
 $$w_t = w_{\text{prev}} - \text{learning_rate} \cdot \frac{\hat{m}_t}{\sqrt{\hat{v}_t} + \epsilon}$$
- Traduzindo para cdigo (Python):
```python
w_t = w_prev - learning_rate * m_t_hat / (sqrt(v_t_hat) + epsilon)
```

Onde:
- `w_t` ($w_t$) so os pesos atualizados.
- `v_t_hat` ($\hat{v}_t$) representa o momento do gradiente ao quadrado corrigido.
- `w_prev` ($w_{\text{prev}}$) so os pesos anteriores.
- `learning_rate`  a taxa de aprendizado.
- `epsilon` ($\epsilon$)  um pequeno valor para evitar diviso por zero.

<hr>

O 'adam' combina o momento do gradiente (para acelerar a convergncia em direo ao mnimo) e o momento do gradiente ao quadrado (para ajustar a taxa de aprendizado de forma adaptativa para cada peso). Essa combinao ajuda a lidar com problemas de superfcie de otimizao complexa, permitindo que o algoritmo se adapte a diferentes escalas de gradiente em diferentes direes.

O 'adam'  amplamente usado devido  sua eficincia e boa performance em muitos cenrios de treinamento de redes neurais. No entanto,  importante ajustar os hiperparmetros do 'adam', como `learning_rate`, `beta_1` e `beta_2`, para obter o melhor desempenho no seu problema especfico.
